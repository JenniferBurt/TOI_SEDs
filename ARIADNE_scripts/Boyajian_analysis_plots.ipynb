```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#csv summarizing parameters of calibration stars derived from ariadne + literature
data = pd.read_csv('CalibrationStars.csv')
```

Radius (includes gaia photometry):

```python
#ariadne radii
ariadne_rad = data['radius']
ariadne_rad_err_up = data['e_ariadne_upper']
ariadne_rad_err_low = data['e_ariadne_lower']

#published radii from Boyajian papers
boyajian = data['published_radius']
boyajian_err = data['pub_e_rad']
```

Metallicity:

```python
#ariadne [Fe/H]
ariadne_feh = data['[Fe/H]']
ariadne_feh_e_up = data['ariadne_feh_upper']
ariadne_feh_e_low = data['ariadne_feh_lower']

#published [Fe/H]
pub_feh = data['published [Fe/H]']
pub_feh_err = data['pub_e_feh']
```

logg:

```python
#ariadne logg
ariadne_logg = data['logg']
ariadne_logg_upper = data['ariadne_logg_upper']
ariadne_logg_lower = data['ariadne_logg_lower']

#published logg
pub_logg = data['published_logg']
pub_logg_err = data['pub_e_logg']
```

Teff:

```python
#ariadne teff
ariadne_teff = data['teff']
ariadne_teff_upper = data['ariadne_teff_upper']
ariadne_teff_lower = data['ariadne_teff_lower']

#published teff
pub_teff = data['published_teff']
pub_teff_err = data['pub_e_teff']
```

Plotting:

```python
fig, axs = plt.subplots(2,2, figsize=(10,10))

#radius
axs[0,0].plot([0.4,1.8],[0.4,1.8],'--')
axs[0,0].errorbar(ariadne_rad, boyajian, xerr=[ariadne_rad_err_up,ariadne_rad_err_low], yerr=boyajian_err, fmt='o')
axs[0,0].set_title('Radius')
axs[0,0].set(xlabel='ARIADNE Values', ylabel='Boyajian values')

#metallicity
axs[0,1].plot([-.2,.8],[-.2,.8],'--')
axs[0,1].errorbar(ariadne_feh, pub_feh, xerr=[ariadne_feh_e_up,ariadne_feh_e_low], yerr=pub_feh_err, fmt='o')
axs[0,1].set(xlabel='ARIADNE values', ylabel='Published Values')
axs[0,1].set_title('[Fe/H]')

#logg
axs[1,0].plot([4,6],[4,6],'--')
axs[1,0].errorbar(ariadne_logg, pub_logg, xerr=[ariadne_logg_upper,ariadne_logg_lower], yerr=pub_logg_err, fmt='o')
axs[1,0].set(xlabel='ARIADNE values', ylabel='Published Values')
axs[1,0].set_title('Log g')

#teff
axs[1,1].plot([3000,7000],[3000,7000],'--')
axs[1,1].errorbar(ariadne_teff, pub_teff, xerr=[ariadne_teff_upper,ariadne_teff_lower], yerr=pub_teff_err, fmt='o')
axs[1,1].set(xlabel='ARIADNE values', ylabel= 'Published values')
axs[1,1].set_title('Teff')

plt.show()
```

ARIADNE outputs after deleting gaia photometry:

Radius:

```python
#ariadne radii
ariadne_ng_rad = data['radius_ng']
ariadne_ng_rad_err_up = data['e_ariadne_ng_upper']
ariadne_ng_rad_err_low = data['e_ariadne_ng_lower']
```

Metallicity:

```python
#ariadne [Fe/H]
ariadne_ng_feh = data['[Fe/H]_ng']
ariadne_ng_feh_e_up = data['ariadne_ng_feh_upper']
ariadne_ng_feh_e_low = data['ariadne_ng_feh_lower']
```

Log g:

```python
#ariadne logg
ariadne_ng_logg = data['logg_ng']
ariadne_ng_logg_upper = data['ariadne_ng_logg_upper']
ariadne_ng_logg_lower = data['ariadne_ng_logg_lower']
```

Teff:

```python
#ariadne teff
ariadne_ng_teff = data['teff_ng']
ariadne_ng_teff_upper = data['ariadne_ng_teff_upper']
ariadne_ng_teff_lower = data['ariadne_ng_teff_lower']
```

Plotting:

```python
fig, axs = plt.subplots(2,2, figsize=(10,10))

#radius
axs[0,0].plot([0.4,1.8],[0.4,1.8],'--')
axs[0,0].errorbar(ariadne_ng_rad, boyajian, xerr=[ariadne_ng_rad_err_up,ariadne_ng_rad_err_low], yerr=boyajian_err, fmt='o', color='mediumseagreen')
axs[0,0].set_title('Radius')
axs[0,0].set(xlabel='ARIADNE Values', ylabel='Boyajian values')

#metallicity
axs[0,1].plot([-.2,.8],[-.2,.8],'--')
axs[0,1].errorbar(ariadne_ng_feh, pub_feh, xerr=[ariadne_ng_feh_e_up,ariadne_ng_feh_e_low], yerr=pub_feh_err, fmt='o',color='mediumseagreen')
axs[0,1].set(xlabel='ARIADNE values', ylabel='Published Values')
axs[0,1].set_title('[Fe/H]')

#logg
axs[1,0].plot([4,6],[4,6],'--')
axs[1,0].errorbar(ariadne_ng_logg, pub_logg, xerr=[ariadne_ng_logg_upper,ariadne_ng_logg_lower], yerr=pub_logg_err, fmt='o',color='mediumseagreen')
axs[1,0].set(xlabel='ARIADNE values', ylabel='Published Values')
axs[1,0].set_title('Log g')

#teff
axs[1,1].plot([3000,7000],[3000,7000],'--')
axs[1,1].errorbar(ariadne_ng_teff, pub_teff, xerr=[ariadne_ng_teff_upper,ariadne_ng_teff_lower], yerr=pub_teff_err, fmt='o',color='mediumseagreen')
axs[1,1].set(xlabel='ARIADNE values', ylabel= 'Published values')
axs[1,1].set_title('Teff')

plt.show()
```

```python
#plot showing outputs both with and without using gaia photometry

fig, axs = plt.subplots(2,2, figsize=(10,10))

#radius
axs[0,0].plot([0.4,1.8],[0.4,1.8],'--')
axs[0,0].errorbar(ariadne_rad, boyajian, xerr=[ariadne_rad_err_up,ariadne_rad_err_low], yerr=boyajian_err, fmt='o', alpha=.5,label = 'With Gaia')
axs[0,0].errorbar(ariadne_ng_rad, boyajian, xerr=[ariadne_ng_rad_err_up,ariadne_ng_rad_err_low], yerr=boyajian_err, fmt='o', alpha=.5, label = 'Without Gaia')
axs[0,0].set_title('Radius')
axs[0,0].set(xlabel='ARIADNE Values', ylabel='Boyajian values')
axs[0,0].legend()

#metallicity
axs[0,1].plot([-.2,.8],[-.2,.8],'--')
axs[0,1].errorbar(ariadne_feh, pub_feh, xerr=[ariadne_feh_e_up,ariadne_feh_e_low], yerr=pub_feh_err, fmt='o', alpha=.5, label='With Gaia')
axs[0,1].errorbar(ariadne_ng_feh, pub_feh, xerr=[ariadne_ng_feh_e_up,ariadne_ng_feh_e_low], yerr=pub_feh_err, fmt='o', alpha=.5,label='Without Gaia')
axs[0,1].set(xlabel='ARIADNE values', ylabel='Published Values')
axs[0,1].set_title('[Fe/H]')
axs[0,1].legend()

#logg
axs[1,0].plot([4,6],[4,6],'--')
axs[1,0].errorbar(ariadne_logg, pub_logg, xerr=[ariadne_logg_upper,ariadne_logg_lower], yerr=pub_logg_err, fmt='o', alpha=.5, label='With Gaia')
axs[1,0].errorbar(ariadne_ng_logg, pub_logg, xerr=[ariadne_ng_logg_upper,ariadne_ng_logg_lower], yerr=pub_logg_err, fmt='o', alpha=.5, label='Without Gaia')
axs[1,0].set(xlabel='ARIADNE values', ylabel='Published Values')
axs[1,0].set_title('Log g')
axs[1,0].legend()

#teff
axs[1,1].plot([3000,7000],[3000,7000],'--')
axs[1,1].errorbar(ariadne_teff, pub_teff, xerr=[ariadne_teff_upper,ariadne_teff_lower], yerr=pub_teff_err, fmt='o', alpha=.5, label='With Gaia')
axs[1,1].errorbar(ariadne_ng_teff, pub_teff, xerr=[ariadne_ng_teff_upper,ariadne_ng_teff_lower], yerr=pub_teff_err, fmt='o', alpha=.5, label='Without Gaia')
axs[1,1].set(xlabel='ARIADNE values', ylabel= 'Published values')
axs[1,1].set_title('Teff')
axs[1,1].legend()

plt.show()
```
