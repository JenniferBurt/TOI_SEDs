```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#radius measurements & errors
ariadne = [1.0188, 0.6194, .7049]
boyajian = [1.166, .5673, .6415]
ariadne_err = [.0639, 0.0334, .0363]
boyajian_err = [.026, .0137, .0048]
```

```python
#plotting ariadne vs boyajian radii
plt.errorbar(x=ariadne[0], y=boyajian[0], xerr=ariadne_err[0], 
yerr=boyajian_err[0], marker='o', label = '18 Sco')

plt.errorbar(x=ariadne[1], y=boyajian[1], xerr=ariadne_err[1], yerr=boyajian_err[1], marker='o', label = 'GJ 338B')

plt.errorbar(x=ariadne[2], y=boyajian[2], xerr=ariadne_err[2], yerr=boyajian_err[2], marker='o', label = 'GJ 380')


plt.plot([0,1.2],[0,1.2],'--') #1:1 line for reference
plt.title('ARIADNE vs Boyajian (2012) Radii')
plt.legend()
plt.show()
```

```python
#teff measurements and errors
ariadne_teff = [5808.7877, 3872.4322, 3995.8850]
ariadne_teff_err = [168.6884, 101.6488, 83.5782]

boyajian_teff = [5433, 3867, 4081]
boyajian_teff_err = [69, 37, 15]
```
