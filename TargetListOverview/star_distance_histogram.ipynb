```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating histogram of stellar distances from TOI list

distance = list(np.loadtxt('toi_distances.txt'))
```

```python
plt.hist(distance, bins=75)
xmin = 0
xmax= 400
plt.xlim(xmin,xmax)
plt.xlabel('Distance (pc)')
plt.ylabel('Counts')
plt.title('Histogram of Stellar Distances')

plt.show()
```
