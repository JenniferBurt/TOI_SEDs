```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Cycle3_TOI_List.csv')

feh = data['metallicity'] #metallicities of 61 out of 136 TOIs

plt.hist(feh, bins=25)
xmin = -.4
xmax= .4
plt.xlim(xmin,xmax)
plt.yticks(np.arange(0,46,5))
plt.xlabel('[Fe/H] (dex)')
plt.ylabel('Counts')
plt.title('Histogram of Stellar Metallicities')

plt.show()
```

```python
x = 0
for i in range(len(feh)):
    if np.isnan(feh[i]) == True:
        x+=1
    else:
        pass

print(136-x) #size minus nan values to find out how many stars have metallicities listed on vizier/exofop
```
