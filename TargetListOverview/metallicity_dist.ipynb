```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Cycle3_TOI_List.csv')

feh = data['metallicity']

plt.hist(feh, bins=25)
xmin = -.4
xmax= .4
plt.xlim(xmin,xmax)
plt.xlabel('[Fe/H] (dex)')
plt.ylabel('Counts')
plt.title('Histogram of Stellar Metallicities')

plt.show()
```
